```python
import tweepy 
import csv
import json
import time 
import re

consumer_key = 'NjC1h1fY9sSCRqfZ5Q8MzUcg7'
consumer_secret= 'aT1SAQ5xhF4mWNS5NBO3ZSzqjOXw45iO2fE5DBXJ2MPWy5cOCd'
access_token = '222166346-dskMUbdOgHeOJ48qCXM64fl9ynDcSPUYJ4uWEUh9'
access_token_secret = 'hjdnzK21RbRFRCJgG4LBU3Ysta50m1XZTzryRlI0mKXdt'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, 
                 wait_on_rate_limit_notify=True, 
                 retry_count=10, retry_delay=5, retry_errors=5)

tweetFile = 'tweetFile '+time.strftime('%x') +'.csv'

with open(r'tweetFile-'+ time.strftime('%d') + '-'+ time.strftime('%b')+ '-'+ time.strftime('%y')+'.csv', 'w', newline='', 
          encoding='utf-8') as csvFile:
    csvWriter = csv.writer(csvFile)
    csvWriter.writerow(['Original Tweet', 'tweet', 'Split Tweet'])

trackers = ['#FTSE','#FTSE100','$III','$ADM','$AAL','$ANTO',
            '$AHT','$ABF','$AZN','$AV.','$BAB','$BA.','$BARC',
            '$BDEV','$BKG','$BLT','$BP.','$BATS','$BLND','$BT.A',
            '$BNZL','$BRBY','$CCL','$CNA','$CCH','$CPG','$CTEC',
            '$CRH','$CRDA','$DCC','$DGE','$DLG','$EZJ','$EXPN',
            '$FERG','$FRES','$GFS','$GKN','$GSK','$GLEN','$HMSO',
            '$HL.','$HSBA','$IMB','$INF','$IHG','$IAG','$ITRK',
            '$ITV','$JMAT','$KGF','$LAND','$LGEN','$LLOY','$LSE',
            '$MKS','$MDC','$MERL','$MCRO','$MNDI','$MRW','$NG.',
            '$NXT','$NMC','$OML','$PPB','$PSON','$PSN','$PRU',
            '$RRS','$RB.','$REL','$RTO','$RIO','$RR.','$RBS',
            '$RDSA','$RSA','$SGE','$SBRY','$SDR','$SMT','$SGRO',
            '$SVT','$SHP','$SKY','$SN.','$SMIN','$SKG','$SSE',
            '$STAN','$SLA','$STJ','$TW.','$TSCO','$TUI','$ULVR',
            '$UU.','$VOD','$WTB','$WPG','$WPP']

def cleanTweet( Tweet ):
    #remove http link 
    new = re.sub(r'http.?:\/\/[^\s]+[\s]?', '', Tweet)
    #remove numbers,hashtags,cashtags and usernames
    new = re.sub(r'((\$|\#|\@)\S*)|\d+', '', new)
    #remove all other punctuation
    cleanTweet = re.sub(r'\!|\"|\#|\$|\%|\&|\(|\)|\*|\+|\,|\-|\.|\/|\:|\;|\<|\=|\>|\?|\^|\_|\`|\~|\\|\[|\]|\{|\}|\|', 
                         '', new)
    return( cleanTweet );
    
def splitUpTweet( Tweet ):
    newTweet = Tweet.split()
    return newTweet;

class twitterListener(tweepy.StreamListener):
    print('Connected to twitter stream')
    print('Searching for following trackers ',trackers)
    
    def on_data(self, data):
        tweetData = json.loads(data)
        tweet = cleanTweet(tweetData['text'])
        print(tweetData['text'])
        splitTweet = splitUpTweet(tweet)
        with open(r'tweetFile-'+ time.strftime('%d') + '-'+ time.strftime('%b')+ '-'+ time.strftime('%y')+'.csv', 'a', 
                   newline='', encoding='utf-8') as csvFile:
            csvWriter.writerow([tweetData['text'], tweet, splitTweet])
        return True
    
    def on_error(self, status):
        print('problem with twitter stream: ',status)
        return False
    
def startStream():
    while True:
        try:
            twitter_stream = tweepy.Stream(api.auth, twitterListener())
            twitter_stream.filter(track=trackers, languages=['en'])
        except:
            continue
            
startStream()
```
